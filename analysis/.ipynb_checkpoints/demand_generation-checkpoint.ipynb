```python
import numpy as np 
import pandas as pd 
import plotly.express as px
import enlighten
import matplotlib.pyplot as plt
import swifter
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
from scipy.spatial import distance
from tqdm import tqdm
```

## we define constants

```python
#variation time to walk to station
MU_WALK=7 #min
SIGMA_WALK=3 #min
```

## we load the data

```python
stops_data=pd.read_xml('./../2021-10-22-19-57-50/stopinfos.xml')
location_data=pd.read_xml('./../2021-10-22-19-57-50/osm_bbox.osm.xml')
uber_pickups=pd.read_csv('./../demand/uber-raw-data-aug14.csv')
map_boundaries={'NE':(40.767318,-73.964559),'NW':(40.768943,-74.008087),'SE':(40.734474,-73.966660),'SW':(40.735069,-74.013205)}
```

## we format the date time data to datetime series and filter points insider boundaries

```python
uber_pickups['Date/Time'] = pd.to_datetime(uber_pickups['Date/Time'])
boundaries=[(40.735069,-74.013205),(40.768943,-74.008087),(40.767318,-73.964559),(40.734474,-73.966660)]
polygon = Polygon(boundaries)
uber_pickups_filtered=uber_pickups.loc[uber_pickups.apply(lambda x: polygon.contains(Point(x.Lat, x.Lon)),axis=1)]
uber_pickups_filtered=uber_pickups_filtered.reset_index()
uber_pickups_filtered.head()
```

## Now we plot the pickup locations

```python
fig = px.scatter_geo(uber_pickups_filtered.iloc[:700],lat='Lat',lon='Lon', hover_name="Date/Time", opacity=0.5)
fig.update_layout(title = 'European map', title_x=0.5)
fig.show()
```

## We join the data for every public transport stop

```python
location_data_filtered = location_data[np.logical_and((location_data['id'].notna()), (location_data['lat'].notna()))]
stops_data['bus_stop_corrected']=pd.to_numeric(stops_data.busStop.map(lambda z: z.split('@')[0]))
join_data_by_station=location_data_filtered.set_index('id').join(stops_data.set_index('bus_stop_corrected'), how='inner')
join_data_by_station=join_data_by_station.reset_index().rename(columns={'index': 'busStop'})
```

```python
fig = px.scatter_geo(join_data_by_station.loc[join_data_by_station.type=='pt_subway'],lat='lat',lon='lon', hover_name="type", opacity=0.5)
fig.update_layout(title = 'New York map', title_x=0.5)
fig.show()
```

## Now we generate the data to move to the trains 

```python
# first we compute and find the closest bus station 
# this is not optimized but it does work 
def arg_min_bus_stop(pickup_point):
    return bus_stations.iloc[bus_stations.apply(lambda stops: distance.euclidean([pickup_point.Lat, pickup_point.Lon], [stops.lat, stops.lon]),axis=1).argmin()]['id']
#now lets run it 
tqdm.pandas()
bus_stations=join_data_by_station.loc[join_data_by_station.type=='pt_bus']
uber_pickups_filtered['id_busStop']=uber_pickups_filtered.apply(arg_min_bus_stop,axis=1)
```

```python
train_stations=join_data_by_station.loc[np.logical_or(join_data_by_station.type=='pt_train',join_data_by_station.type=='pt_subway')]
uber_pickups_filtered['id_TrainStation']=train_stations.iloc[np.random.randint(train_stations.shape[0], size=uber_pickups_filtered.shape[0])].reset_index()['id']
uber_pickups_filtered['time_for_walk']=uber_pickups_filtered['Date/Time']
uber_pickups_filtered['time_for_bus']=uber_pickups_filtered['Date/Time']+ pd.to_timedelta(np.random.normal(MU_WALK, SIGMA_WALK, uber_pickups_filtered['Date/Time'].shape[0]),unit='m')
```

```python
map_bound_try=[(40.735069,-74.013205),(40.768943,-74.008087),(40.767318,-73.964559),(40.734474,-73.966660)]
polygon = Polygon(map_bound_try)
plt.plot(*polygon.exterior.xy)
point = Point(40.75059649866779, -73.9933677707311)
print(polygon.contains(point))
```
