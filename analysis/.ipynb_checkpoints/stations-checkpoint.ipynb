```python
import numpy as np 
import pandas as pd 
import plotly.express as px
import enlighten
import matplotlib.pyplot as plt
```

```python
stations_data=pd.read_csv('./../stations/stations.csv', sep =';')
traffic_data=pd.read_excel('./../stations/station_usage_estimates_1112.xlsx', sheet_name='Estimates of station usage 1112')
```

```python
stations_data['color']=stations_data.is_main_station.map(lambda x: 10 if x=='t' else 1)
traffic_data['station_name_norm']=traffic_data['Station Name'].map(lambda x : (x.lower()).replace(' ','-'))
```

```python
uk_stations=stations_data.loc[(stations_data.country=='GB') & (stations_data.color==10)].sort_values('color')
join_data_by_station=traffic_data.set_index('station_name_norm').join(uk_stations.set_index('slug'), how='inner')
join_data_by_station=join_data_by_station.reset_index().rename(columns={'index': 'slug'})
```

```python
fig = px.scatter_geo(join_data_by_station,lat='latitude',lon='longitude', hover_name="slug", scope='europe', color='color', opacity=0.5)
fig.update_layout(title = 'European map', title_x=0.5)
fig.show()
```

```python
def generate_travel_samples(join_data_by_station):
    travels=[]
    stations=[slug for slug in join_data_by_station.slug]
    travels_counts={join_data_by_station.slug.iloc[row]:
                    [0,join_data_by_station['Entries Total'].iloc[row]//12,0,join_data_by_station['Exits Total'].iloc[row]//12]
                    for row in range(join_data_by_station.shape[0])} #(entries,target entries, exits, target exits)
    total_travels=0
    target_travels=join_data_by_station['Entries Total'].sum()//12
    manager = enlighten.get_manager()
    travels_bar = manager.counter(total=int(target_travels), desc='travels', unit='travel')
    while total_travels<target_travels and len(stations)>1:
        origin=stations[np.random.randint(0,len(stations))]
        destiny=stations[np.random.randint(0,len(stations))]          
        if origin==destiny:
            continue 
        if travels_counts[origin][0]<travels_counts[origin][1] and travels_counts[destiny][2]<travels_counts[destiny][3]:
            travels.append((origin, destiny,np.random.randint(1,24)))
            travels_counts[origin][0]+=1
            travels_counts[destiny][2]+=1
            travels_bar.update()
            total_travels+=1
            if((travels_counts[origin][0])>(travels_counts[origin][1]-1) and travels_counts[origin][2]>(travels_counts[origin][3]-1)):
                stations.remove(origin)
            if(travels_counts[destiny][0]>(travels_counts[destiny][1]-1) and travels_counts[destiny][2]>(travels_counts[destiny][3]-1)):
                stations.remove(destiny)
    manager.stop()
    print("unable to fullfil demand for {} by {} travels".format(stations[0],travels_counts[stations[0]][1]-travels_counts[stations[0]][0]))
    return travels
    
```
