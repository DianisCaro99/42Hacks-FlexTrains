```python
import numpy as np 
import pandas as pd
import bs4
import lxml
from bs4 import BeautifulSoup as bs
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

## Now we statate the initial parameters for the base scenario

```python
#every agent will have a tuple with the following factors (Through traffic factor, Count )
#Through traffic factor: defines how many times it is more likely for an edge at the boundary of the simulation area being chosen compared to an edge entirely located inside the simulation area
#Count: parameter defines how many vehicles are generated per hour and lane-kilometer.
#Initial relation between cars and rails usage is 83/8
initial_parameters={'cars':(2,51),'pederestians':(2,5),'bus':(5,4),'urban trains':(20,1),'trains':(20,1)}
```

## We extract the relevant information for the base case scenario

### We start with emissions

```python
#First we define a function to read XML file properly 
# Read the XML file
bus_emissions={'CO2':[],'PM':[]}
car_emissions={'CO2':[],'PM':[]}
with open('./../data_for_stations/statistics/emission.xml', "r") as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    for line in content:
        if('<timestep' in line):
            bus_emissions['CO2'].append(0)
            bus_emissions['PM'].append(0)
            car_emissions['CO2'].append(0)
            car_emissions['PM'].append(0)
        else:
            if('type="pt_bus"' in line):
                bus_emissions['CO2'][-1]+=float(list(filter(lambda x: 'CO2='in x,line.split(' ')))[0].split('"')[-2])
                bus_emissions['PM'][-1]+=float(list(filter(lambda x: 'PMx='in x,line.split(' ')))[0].split('"')[-2])
            elif('type="veh_passenger"' in line):
                car_emissions['CO2'][-1]+=float(list(filter(lambda x: 'CO2='in x,line.split(' ')))[0].split('"')[-2])
                car_emissions['PM'][-1]+=float(list(filter(lambda x: 'PMx='in x,line.split(' ')))[0].split('"')[-2])
```

```python
bus=pd.DataFrame(bus_emissions)
car=pd.DataFrame(car_emissions)
```

```python
ax =sns.scatterplot(bus.index, bus['CO2'], label="bus")
sns.scatterplot(car.index, car['CO2'], label="car")
ax.set(xlabel="time_step", ylabel="CO2 emissions")
```

```python
ax =sns.scatterplot(bus.index, bus['PM'], label="bus")
sns.scatterplot(car.index, car['PM'], label="car")
ax.set(xlabel="time_step", ylabel="PM emissions")
```

```python
print('Busses emitted a total of {} CO2 emissions'.format(round(bus.CO2.sum(),1)))
print('Cars emitted a total of {} CO2 emissions'.format(round(car.CO2.sum(),1)))
print('Cars emitted {} times more CO2 emissions than busses'.format(round(car.CO2.sum()/bus.CO2.sum(),1)))
```

### We move to trip statistics

```python
#First we define a function to read XML file properly 
# Read the XML file
with open('./../data_for_stations/trip-statistic.xml', "r") as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    trip_content = bs(content, "lxml")
```

```python
vel_bus=float(trip_content.find('ridestatistics')['routelength'])/float(trip_content.find('ridestatistics')['duration'])
vel_car=float(trip_content.find('vehicletripstatistics')['routelength'])/float(trip_content.find('vehicletripstatistics')['duration'])
print('Every bus ride took on average {} steps per unit of distance.'.format(vel_bus))
print('Every car ride took on average {} steps per unit of distance.'.format(vel_car))
print('car rides were {} times faster than bus rides.'.format(round(vel_car/vel_bus,1)))
```

### And trip info

```python
#First we define a function to read XML file properly 
# Read the XML file
with open('./../data_for_stations/trip.xml', "r") as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    trips_content = bs(content, "lxml")
```

```python
bus_trips={'duration':[]}
car_trips={'duration':[]}
for vehicle in trips_content.find_all("tripinfo"):
    if(vehicle.get('vtype')=='urban_rail_urban'):
        bus_trips['duration'].append(float(vehicle.get('duration')))
    elif(vehicle.get('vtype')=='veh_passenger'):         
        car_trips['duration'].append(float(vehicle.get('duration')))
bus_trips_pd=pd.DataFrame(bus_trips)
car_trips_pd=pd.DataFrame(car_trips)
```
